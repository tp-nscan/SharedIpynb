```python
%matplotlib inline
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

### Beta distribution sampler

```python
bdist = np.random.beta(0.1, 0.2, size=20000)
bins = np.arange(0, 1.01, 0.01)
plt.figure(figsize=(6, 2))
plt.title
plt.hist(bdist, bins=bins)
plt.show()
```

```python
pdf, bin_edges = np.histogram(bdist, bins=bins, density=True)
plt.figure(figsize=(6, 3))
plt.plot(pdf)
plt.plot(np.cumsum(pdf))
plt.legend(['pdf', 'cdf'], ncol=2, loc='upper left');
plt.show()
```

### Tanh

```python
dom = np.linspace(-3, 3, 100)
plt.figure(figsize=(6, 2))
plt.plot(dom, np.tanh(dom))
plt.legend(['tanh'], ncol=1, loc='upper left');
plt.show()
```

### Random walk

```python
walkLength = 1000
numwalks = 10
```

```python
dom = np.linspace(0, walkLength, walkLength)
plt.figure(figsize=(10, 2))
for i in range(numwalks):
    plt.plot(dom, np.cumsum(np.random.uniform(-0.5, 0.5, walkLength)))
plt.show()
```

### Gamma Distribution

```python
dom = np.linspace(0.05, 5, 1000)
plt.figure(figsize=(6, 2))
plt.plot(dom, scp.special.gamma(dom))
plt.title('gamma');
plt.show()
```

### Poisson Distribution

```python
def Poisson(l:float, x:float):
    return np.exp(-l) * np.power(l, x) / scp.special.gamma(x + 1)
```

```python
dom = np.linspace(0.01, 20, 1000)
plt.figure(figsize=(6, 2))
plt.plot(dom, Poisson(2., dom), label='l=2')
plt.plot(dom, Poisson(3.5, dom), label='l=3.5')
plt.plot(dom, Poisson(6., dom), label='l=6')
plt.plot(dom, Poisson(12., dom), label='l=12')
#plt.axis('equal')
plt.legend();
plt.title('Poisson');
plt.show()
```

```python
dom = np.linspace(0.0001, 10, 1000)
plt.figure(figsize=(6, 2))
plt.plot(dom, np.sin(dom) / dom)
plt.title('sin(x)/x');
plt.show()
```
